```python
#Partitioning and repartitioning
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, desc, sum, avg
# Initialize
spark = SparkSession.builder.appName("SparkCompleteNotes").getOrCreate()
# Create Base DataFrame
```

```python
df = spark.range(0,1000000).withColumn("value", col("id")%1000)
```

```python
df.take(10)
```

```python
print("Before Partition:" , df.rdd.getNumPartitions())
df.write.mode("overwrite").csv("output/BeforeEmp" , header = True)
```

```python
df_repartition=df.repartition(20)
print("After Partition:", df_repartition.rdd.getNumPartitions())
```

```python
df_repartition.write.mode("overwrite").csv("output/afterEmp" , header = True)
```

```python
df_coal = df_repartition.coalesce(2)
print("After coalesced:" , df_coal.rdd.getNumPartitions())
```
